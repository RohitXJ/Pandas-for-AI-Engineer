# Findings
* There are 100k rows and 8 columns in this dataset.
* Mismatch of dtypes for multiple columns.
* The should be numerical columns cannot be easily converted due the presence of string values in it, i.e 'Price' column where in some places the $ sign in present, some price values are even in negative, this also needs to be taken care of.
* Solutin for 'Price' column : After we remove the $ character and fix the negative sign issues, to fill the NaN gap we can find out the average unit price of each product categroy and then put those, then we will be able to bring back most of the data, and i feel like it's better than putting any all-over mean or average values in there, since the quantity is also a factor not to be ignored.
* The 'Quantity' column too has the negative sign problem along with some null data.
* Solution for 'Quantity' column : We can use the same method as of 'Price' column, by finding the average quantity value for each product type and imputing by that.
* "Product_Name" column has the problem of incorrect or incomplete names of products, this can be solved by using the Regular Expression lib.
* Column "Payment_Method" and "Transaction_Status" have the same naming problem as "Product_Name", and it can be solved using the regular expression lib, better convert the texts to lowercase for uniformity. It would be better to put new features based on payment methods, like putting short forms i.e "CC" instead of "credit card", "C" instead of "cash", etc. same for the "Payment_Status"" column, its better to label encode the status.
* "Transaction_Date" has a problem with incorrect date format and missing data, we can convert it into datetime format and fill the missing values by forward filling or backward filling.
* "Transaction_ID" and "Customer_ID" columns can be dropped if this data is going to be fed into any ML model, as they are not required for prediction, but if this data is for any sales team then these columns should be kept for unsupervised prediction purpose.
* After the cleaning of the dataset, the clean data must be label encoded (since there are more columns with categorical values) and scaled(price and quantity number digits vary a lot).
