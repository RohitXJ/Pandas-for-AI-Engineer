# Imports

```python
import pandas as pd
import numpy as np
```

# Data Import

```python
df = pd.read_csv("dirty_cafe_sales.csv")
df.head()
```

```python
df.columns = df.columns.str.lower().str.replace(" ", "_")
df = df.drop(columns="transaction_id")  # Drop transaction_id column as it is not needed asuming this data will be fed into some ML model
df.head()
```

```python
df.info(memory_usage="deep")
```

```python
print(f"Missing Value Counts\n{df.isnull().sum()}\n")
print(f"Missing Value Counts : Percentage\n{df.isnull().sum() / len(df) * 100}\n")
```

```python
print(f"Total Missing Value Counts\n{df.isnull().sum().sum()} Out of {len(df)*len(df.columns)}\n")
print(f"Total Missing Value Counts : Percentage : {df.isnull().sum().sum() / (len(df)*len(df.columns)) * 100}\n")
```

# Null conversion

```python
df.sample(10)
```

```python
df.columns
```

```python
error_set = ['UNKNOWN', 'ERROR', None]
```

```python
df = df.replace(error_set, np.nan)
```

```python
df[["quantity","price_per_unit","total_spent"]] = df[["quantity","price_per_unit","total_spent"]].astype("float32")
```

```python
df.sample(10)
```

```python
print(f"Missing Value Counts\n{df.isnull().sum()}\n")
```

All unwanted values are converted to NaN for better access and simplification of data processing process.

# Recovering `quantity`, `price_per_unit` and `total_spent` column data 

```python
df['quantity'] = df['quantity'].fillna(df['total_spent'] // df['price_per_unit'])

df['price_per_unit'] = df['price_per_unit'].fillna(df['total_spent'] / df['quantity'])

df['total_spent'] = df['total_spent'].fillna(df['quantity'] * df['price_per_unit'])
```

```python
df.sample(10)
```

```python
print(f"Missing Value Counts\n{df.isnull().sum()}\n")
```

# Recovering `item` column data from previously recovered columns

```python
price_map = df.drop_duplicates('item').set_index('item')['price_per_unit'].to_dict()
del price_map[np.nan]
price_map
```

```python
item_lookup = {v: k for k, v in price_map.items() if pd.notna(k)}
item_lookup
```

```python
df['item'] = df['item'].fillna(df['price_per_unit'].map(item_lookup))
```

```python
print(f"Missing Value Counts\n{df.isnull().sum()}\n")
```

We can see a drastic change in the number of recovered data, from 969 missing data in `item` column to now 6. <br>
The 2-4no. columns could be again checked for data recovery but that part is being skipped and the rows will be dropped for simplicity, since that is very little data to waste time on.

# Imputing the last three columns

```python
df['payment_method'] = df['payment_method'].fillna(df['payment_method'].mode()[0])
df['location'] = df['location'].fillna(df['location'].mode()[0])
df['transaction_date'] = df['transaction_date'].ffill() 
df['transaction_date'] = pd.to_datetime(df['transaction_date'])
```

```python
df.dropna(axis="rows", inplace=True)
# Converting the last columns datatype and sorting based on that column
df = df.sort_values(by='transaction_date', ascending=True)
df = df.reset_index(drop=True)
df['quantity'] = df['quantity'].astype("int16")
df.info(memory_usage="deep")
```

```python
df.sample(10)
```

# Exporting Data

```python
df.to_csv("cleaned_cafe_sales_data.csv",index=False)
```
